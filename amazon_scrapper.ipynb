```python
# !pip install beautifulsoup4
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
all_links = []
all_names = []
all_rate = []
all_review = []
all_prices = []

URL = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
r = requests.get(url=URL, headers=headers)
soup = BeautifulSoup(r.content, "html.parser")
datas1 = soup.find_all('div', attrs={'class': 'a-section a-spacing-small a-spacing-top-small'})
for i in datas1[2:]:
    all_names.append(i.find('span', attrs={'class': 'a-size-medium a-color-base a-text-normal'}).text)
    all_prices.append(i.find('span', attrs={'class': 'a-price-whole'}).text)
    all_links.append("https://www.amazon.in/" +i.find('a', attrs={'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'},href=True)['href'])
    all_rate.append(i.find('span', attrs={'class': 'a-icon-alt'}).text[:3])
    all_review.append(i.find('span', attrs={'class': 'a-size-base s-underline-text'}).text)
```
